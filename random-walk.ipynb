# 1 D Random walk

```python
n = 10_000
hit_zero = 0
start = 0

# jump
for i in range(n):
    jump = np.random.choice([1,-1])
    start += jump
    if start == 0:
        hit_zero += 1
print(hit_zero)
```

```python
n = 10_000
hit_zero = 0
dim = 2
start = np.zeros(dim)

# jump
for i in range(n):
    step = np.random.choice([1,-1])
    dimension = np.random.choice(range(dim))
    jump = np.zeros(dim) 
    jump[dimension] = step
    start += jump
    if np.all(start == np.zeros(dim)):
        hit_zero += 1
print(hit_zero)
```

# n dimensional Random walk

```python
def ndrand_walk(n=10000,dim=2):
    hit_zero = 0
#     dim = 2
    start = np.zeros(dim)

    # jump
    for i in range(n):
        step = np.random.choice([1,-1])
        dimension = np.random.choice(range(dim))
        jump = np.zeros(dim) 
        jump[dimension] = step
        start += jump
        if np.all(start == np.zeros(dim)):
            hit_zero += 1
    print(f'In {n} runs in {dim}-dimesional RW, we have hit zeros : {hit_zero} many times')
    
ndrand_walk(10000,4)
```

```python
n = 10000
for d in range(1,10):
    ndrand_walk(n, d)
```
