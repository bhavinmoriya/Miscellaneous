One survey of 500 Penn State University students about their favorite sport to watch shows that 283 said Football, 126 said Basketball, 45 said Hockey, 46 said Others. Practice with Minitab to create a pie chart and bar chart for favorite sport.

```python
labels = 'Football Basketball Hockey Others'.split()
sizes = [283,126,45,46]

fig, ax = plt.subplots(figsize=(20,15))
ax.pie(sizes, labels=labels, autopct='%1.1f%%',
      pctdistance=.5, labeldistance=.6,
      shadow=True);
```

```python
fig, ax = plt.subplots(figsize=(20,15))
ax.bar(height=sizes, x=labels);
```
